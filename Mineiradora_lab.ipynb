```python
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime
import numpy as np
from plotly.subplots import make_subplots
from IPython.display import display,clear_output
import panel as pn

style = {'description_width': 'initial'}
pn.extension('plotly')
```

```python
# ajeitando o tipo do dado pelo formato do import
#df = pd.read_csv('tabuas.txt', header=0)
df=pd.read_csv('https://kzweb.com.br/somar/tabuas.txt', header=0)
df = df.iloc[:,:-1]

for x in ['HORA1','HORA2','HORA3']:
    df[x] = df['DATA'] + ' '+ df[x]
    df[x] = pd.to_datetime(df[x]).dt.time
df['DATA'] =  pd.to_datetime(df['DATA'])
df['ANO_MES'] = df['DATA'].dt.to_period('M')
#retiradas datas anteriores por serem 3 observações de zero em 2018
df = df[df['DATA'].dt.year>=2019]
df = df.iloc[:-30,:]
df.index = df['DATA']
```

## Estações do Ano

```python
%%capture --no-stdout

winterStart = '-06-21'
winterEnd   = '-09-22'
summerStart = '-12-21'
summerEnd   = '-03-21'
autumnStart = '-03-21'
autumnEnd   = '-06-21'
springStart = '-09-23'
springEnd  = '-12-23'

df_winter = pd.DataFrame()
for year in range(2019, 2022):
    df_winter = df_winter.append(df.loc[str(year) + winterStart : str(year) + winterEnd]) 
    # used year and year+1 because winter season spans from an initial year to the next year.

df_summer = pd.DataFrame()
for year in range(2019, 2022):
    df_summer = df_summer.append(df.loc[str(year) + summerStart : str(year+1) + summerEnd])

df_autumn = pd.DataFrame()
for year in range(2019, 2022):
    df_autumn = df_autumn.append(df.loc[str(year) + autumnStart : str(year) + autumnEnd]) 
 

df_spring = pd.DataFrame()
for year in range(2019, 2022):
    df_spring = df_spring.append(df.loc[str(year) + springStart : str(year) + springEnd])

```

```python
localidade = pn.widgets.Select(name='localidade', options=['SOMAR', 'MANGA','PORTO'], value='MANGA')
ano = pn.widgets.Select(name='ano', options=['2019', '2020','2021'], value='2019')
```

```python
@pn.depends(ano.param.value, localidade.param.value)
def update_grafico(ano, localidade):

    df_a = df[df['DATA'].dt.year==int(ano)]

    row_n = np.unique(df_a['DATA'].dt.month)

    sepc=[]

    for x in range(0, len(row_n)): #range(1,13): 
        sepc.append([{}])

    fig = make_subplots(
        rows=len(row_n), 
        cols=1, 
        specs = sepc,
        vertical_spacing=0.04)
         

    for month in np.unique(df_a['DATA'].dt.month):

        df_m = df_a[df_a['DATA'].dt.month==month]

        fig.add_trace(go.Scatter(x=df_m['DATA'],
                         y=df_m[localidade+str(1)], name='7:00',
                         line=dict(color='firebrick', width=4)),row=month, col=1)

        fig.add_trace(go.Scatter(x=df_m['DATA'],
                         y=df_m[localidade+str(2)], name='16:00',
                         line=dict(color='royalblue', width=4)),row=month, col=1)

        if(localidade == 'SOMAR'):
            fig.add_trace(go.Scatter(x=df_m['DATA'],
                             y=df_m[localidade+str(3)], name='23:00',
                             line=dict(color='forestgreen', width=4)), row=month, col=1)

        #fig.layout.annotations[month].update(text=month.strftime("%b"))

    fig.update_layout(title='Medição %s<br><sup>%s</sup>'%(localidade, ano),autosize=False,
    width=1480,
    height=1400,)

    return fig
```

```python
p=pn.interact(update_grafico, ano=ano, localidade=localidade)
column = pn.Column(fig1, fig2, fig3, fig4, fig5, fig6, fig7, fig8, fig9, fig10, fig11, fig12, fig13, fig14, fig15, fig16, fig17, fig18, fig19, fig20, fig21, fig22, fig23, fig24, fig25, fig26, p).servable()
```
